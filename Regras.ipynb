### Base Credit data - 97,80%

```python
base_credit = Orange.data.Table('/content/drive/MyDrive/Machine Learning/Download/Bases de dados/credit_data_regras.csv')
```

```python
base_credit.domain
```

```python
base_dividida = Orange.evaluation.testing.sample(base_credit, n=0.25,random_state=0)
```

```python
base_treinamento = base_dividida[1]
base_teste = base_dividida[0]
```

```python
len(base_treinamento), len(base_teste)
```

```python
cn2_credit = Orange.classification.rules.CN2Learner()
regras_credit = cn2_credit(base_treinamento)
```

```python
for regras in regras_credit.rule_list:
  print(regras)
```

```python
previsoes_credit = Orange.evaluation.testing.TestOnTestData(base_treinamento, base_teste, [lambda testdata: regras_credit])
```

```python
previsoes_credit
```

```python
Orange.evaluation.CA(previsoes_credit)
```
