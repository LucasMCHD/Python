# Regressão Logística

```python
from sklearn.linear_model import LogisticRegression
```

### Base Risco Crédito

```python
import pickle
import numpy as np

with open('/content/drive/MyDrive/Machine Learning/Classificação/risco_credito.pkl', 'rb') as f:
  X_risco_credito, y_risco_credito = pickle.load(f)
```

```python
# Apagar os valores 'moderados'
X_risco_credito = np.delete(X_risco_credito, [2,7,11], axis=0)
y_risco_credito = np.delete(y_risco_credito, [2,7,11], axis=0)
```

```python
logistic_risco_credito = LogisticRegression(random_state=1)
logistic_risco_credito.fit(X_risco_credito, y_risco_credito)
```

```python
logistic_risco_credito.intercept_
```

```python
logistic_risco_credito.coef_
```

```python
previsoes_risco = logistic_risco_credito.predict([[0,0,1,2],[2,0,0,0]])
previsoes_risco
```

### Base Credit

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
import numpy as np
from sklearn.linear_model import LogisticRegression

with open('/content/drive/MyDrive/Machine Learning/Classificação/credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
logistic_credit = LogisticRegression(random_state=0)
logistic_credit.fit(X_credit_treinamento,y_credit_treinamento)
```

```python
logistic_credit.intercept_
```

```python
logistic_credit.coef_
```

```python
previsoes_credit = logistic_credit.predict(X_credit_teste)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_teste,previsoes_credit)
```
